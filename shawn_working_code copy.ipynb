```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
from datetime import datetime, timedelta
import time
```

```python
# load the API key
load_dotenv()
av_api_key = os.getenv("AV_API_KEY") 
```

```python
# define the query date range
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)

print(start_date)
```

```python
# pull the sentiment analysis data for Nvidia, Tesla, Apple and Boeing
df_tickers = pd.DataFrame()
counter = 0
tickers = str("AAPL")

# Iterate through each date
pull_date = start_date
while pull_date <= end_date:
    # Convert to datetime object to the API format
    counter =+ 1
    counter_check = counter % 75

    if counter_check == 0:  
        time.sleep(60)
    
    pull_start = pull_date.strftime("%Y%m%dT%H%M")
    pull_end_dt = pull_date + timedelta(days=1)
    pull_end = pull_end_dt.strftime("%Y%m%dT%H%M")
    
    url = 'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=' + tickers + '&apikey=M9BTYXZI7EG8P00S' + '&time_from=' + pull_start + '&time_to=' + pull_end + '&limit=1000'
    r = requests.get(url)
    data = r.json()

    df_data = pd.json_normalize(data['feed'])
    df_tickers = pd.concat([df_tickers, df_data], ignore_index=True)

    pull_date += timedelta(days=1)

display(df_tickers)
```

```python
# From the json, pull the title, date & time, relevance score, sentiment score, and sentiment label

# Use pd.json_normalize to convert the JSON to a DataFrame
df_test = pd.json_normalize(data['feed'])

# Rename the columns for clarity
#df.columns = ['Title', 'Author_First_Name', 'Author_Last_Name', 'Publisher_Name', 'Publisher_Location']

# Display the DataFrame
display(df_test)
```

```python
# filter the data frame for AAPL
# I probably don't need this step
df_AAPL = df_test[df_test['ticker_sentiment'].apply(lambda x: any(d['ticker'] == 'AAPL' for d in x))]
```

```python
# expand the ticker sentiment column
df_AAPL = df_AAPL.explode('ticker_sentiment')
df_AAPL = df_AAPL[df_AAPL['ticker_sentiment'].apply(lambda x: x['ticker'] == 'AAPL')]
df_AAPL = pd.concat([df_AAPL.drop(['ticker_sentiment'], axis=1), df_AAPL['ticker_sentiment'].apply(pd.Series)], axis=1)
```

```python
# review
df_AAPL = df_AAPL.drop(['banner_image', 'source', 'authors', 'category_within_source', 'topics', 'overall_sentiment_score', 'overall_sentiment_label', 'source_domain'], axis=1)
display(df_AAPL)
```

```python
# convert the date & time to a datetime field

def convert_to_datetime(time_stamp):
    date_format = "%Y%m%dT%H%M%S"
    return datetime.strptime(time_stamp, date_format)

df_AAPL['time_published'] = df_AAPL['time_published'].apply(convert_to_datetime)

display(df_AAPL)
```
